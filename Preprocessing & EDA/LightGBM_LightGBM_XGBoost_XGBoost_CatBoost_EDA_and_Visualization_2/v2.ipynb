```python
import warnings
warnings.simplefilter(action='ignore')

import os
import requests
import joblib

import numpy, pandas
import pandas as pd

import matplotlib.pyplot as plt
import scipy.stats as stats
from catboost import CatBoostRegressor ,Pool
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor

from sklearn.model_selection import train_test_split
```

```python
train_c =pandas.read_parquet('/kaggle/input/drw-crypto-market-prediction/train.parquet',
                           engine='pyarrow').replace([numpy.inf, -numpy.inf], 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_c.drop(columns=['label']), train_c['label'], 
                                                    test_size=0.07, random_state=100)
```

```python
R_Forest_parm = {'n_estimators' : 25, 
                 'min_samples_split' : 2, 
                 'max_depth' : 10, 
                 'min_samples_leaf' : 2, 
                 'random_state' : 42}
```

```python
score = {}
for i in X_train.columns:
    print(f'{i} X label')
    catboost = RandomForestRegressor(**R_Forest_parm)
    catboost.fit(numpy.array(X_train[i].drop(columns=['timestamp'])).reshape(-1, 1), y_train)
    
    score[i] = 1 - mean_squared_error(y_test, catboost.predict(numpy.array(X_test[i].drop(columns=['timestamp'])).reshape(-1, 1)))
    print(f'MSE SCORE : {score[i]}')
```
