```python
import warnings
warnings.simplefilter(action='ignore')

import os
import requests
import joblib

import numpy, pandas
import pandas as pd

import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
train_c =pandas.read_parquet('/kaggle/input/drw-crypto-market-prediction/train.parquet',
                           engine='pyarrow').replace([numpy.inf, -numpy.inf], 0)
test_c =pandas.read_parquet('/kaggle/input/drw-crypto-market-prediction/test.parquet',
                           engine='pyarrow').replace([numpy.inf, -numpy.inf], 0)
```

```python
corrulation = {'columns' : [], 'corrulation' : []}
for i in train_c:
    if i != 'label':
        corrulation['columns'].append(i)
        corrulation['corrulation'].append((train_c['label'].corr(train_c[i])))
```

```python
drw = pandas.DataFrame(corrulation)
```

```python
drw[drw['corrulation'] > 0.04]['columns'].to_numpy()
```

```python
train_c[list(drw[drw['corrulation'] > 0.04]['columns'].to_numpy()) + ['label']].to_csv('train.csv', index = False)
```

```python
test_c[list(drw[drw['corrulation'] > 0.04]['columns'].to_numpy())].to_csv('test.csv', index = False)
```
